## 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate synthetic dataset with correlated features
np.random.seed(0)
x = np.random.randn(100)
y = 0.5 * x + np.random.randn(100)
z = 0.7 * y + np.random.randn(100)

data = np.vstack([x, y, z]).T
columns = ['Feature 1', 'Feature 2', 'Feature 3']

# Compute the correlation matrix
corr_matrix = np.corrcoef(data, rowvar=False)

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, xticklabels=columns, yticklabels=columns, cmap='coolwarm', center=0)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

# Given data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add color bar for reference
fig.colorbar(surf)

# Add title and labels
ax.set_title('3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Given dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))

# Plot data
plt.scatter(df['GDP'], df['Population'], s=df['Population'], alpha=0.5, c='blue', edgecolors='w', linewidth=0.5)

# Add labels
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country, fontsize=12, ha='center', va='center')

# Add title and labels
plt.title('Bubble Chart of Country Population vs GDP')
plt.xlabel('GDP')
plt.ylabel('Population')

# Show plot
plt.grid(True)
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output for the notebook
output_notebook()

# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the plot
show(p)
```

## 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Prepare the output for the notebook
output_notebook()

# Generate random x and y values
np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10

# Generate sizes and colors
sizes = np.random.rand(n) * 20 + 10  # Sizes ranging from 10 to 30
colors = np.random.choice(['red', 'green', 'blue', 'orange'], n)

# Create a Bokeh scatter plot
p = figure(title="Scatter Plot with Random Values", x_axis_label='x', y_axis_label='y')
p.scatter(x, y, size=sizes, color=colors, fill_alpha=0.6, legend_field='color')

# Show the plot
show(p)
```

## 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

#### fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#### counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool

# Prepare the output for the notebook
output_notebook()

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', height=350, width=600)
p.vbar(x=fruits, top=counts, width=0.5, color='blue', alpha=0.7)

# Add labels to the bars
hover = HoverTool()
hover.tooltips = [("Fruit", "@x"), ("Count", "@top")]
p.add_tools(hover)

# Show the plot
show(p)

```

## 4. Create a Bokeh histogram to visualize the distribution of the given data.

#### data_hist = np.random.randn(1000)
#### hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output for the notebook
output_notebook()

# Generate data and compute histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh plot for the histogram
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency', height=400, width=600)

# Add the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black", alpha=0.7)

# Show the plot
show(p)
```

## 5. Create a Bokeh heatmap using the provided dataset

#### data_heatmap = np.random.rand(10, 10)
#### x = np.linspace(0, 1, 10)

#### xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Viridis256
import numpy as np

# Prepare the output for the notebook
output_notebook()

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for Bokeh
data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()

# Create a Bokeh plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', 
           x_range=(0, 1), y_range=(0, 1), width=600, height=600)

# Add the heatmap using image function
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette=Viridis256, level='image')

# Add grid and axis labels
p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_label_orientation = "horizontal"

# Show the plot
show(p)
```
